### STEP TWO: Using Pandas to clean and join recipes with ingredients and add to MongoDB collection.

As we saw in the previous notebook, further cleanup of recipe data and joining of recipes with their corresponding ingredients using Pandas methods. These clean datasets are uploaded into collections in the Mongo database.

```python
# Cleaning up recipe titles and ingredients using Pandas
# Intermediary step before continuing project
#DO NOT RUN THIS CODE FROM HERE!!!!!!!!!!!! SCROLL DOWN

# Dependencies
import pandas as pd
from pymongo import MongoClient
```

```python
# Establish connection to MongoDB
client = MongoClient('mongodb://localhost:27017')
db = client.allrecipes_etl1
collection = db.recipes
collection2= db.ingredients
data = pd.DataFrame(list(collection.find()))
```

```python
# Convert MongoDB collections into dataframes
df = pd.DataFrame(list(collection.find()))
df2 = pd.DataFrame(list(collection2.find()))
```

```python
df
```

```python
df2
```

```python
# Join recipe titles with corresponding ingredients

recipes= df.join(df2, lsuffix='_caller', rsuffix='_other')
recipes.head()
```

```python
# Delete MongoDB id columns
recipes= recipes.drop(columns=['_id_caller', '_id_other'])
```

```python
# RESUME RUNNING CODE FROM HERE
# Read in csv of clean recipe data 
recipes
recipes.to_csv('recipespd.txt',sep='\t')
```

### STEP THREE: Extract FoodData Central dataset and clean data

```python
# Create DataFrame of food database data using Pandas
df = pd.read_excel('https://query.data.world/s/im5ovw5g7a6rkzyjigldmcblimj7jl')
```

```python
# Check DataFrame
df.head()
```

```python
# Remove unwanted columns from DataFrame

organized_df = df[['NDB_No','Shrt_Desc', 'Water_(g)', 'Energ_Kcal', 'Protein_(g)',
       'Lipid_Tot_(g)', 'Carbohydrt_(g)', 'Fiber_TD_(g)',
       'Sugar_Tot_(g)', 'Calcium_(mg)', 'Iron_(mg)','Sodium_(mg)','Vit_C_(mg)','Folate_Tot_(µg)','Choline_Tot_ (mg)', 'Vit_B12_(µg)','Retinol_(µg)','Cholestrl_(mg)', 'GmWt_1', 'GmWt_Desc1', 'GmWt_2',
       'GmWt_Desc2']]
organized_df
```

```python
# Rename columns
organized_df= organized_df.rename(columns={'Shrt_Desc':'Ingredients', 'Lipid_Tot_(g)':'Fat_Tot_(g)'})
```

```python
organized_df.head()
```

```python
# Check number of food data entries
organized_df.count()
```

```python
# Export DataFrame as csv
organized_df.to_csv('foodpd.txt',sep='\t')
```

### STEP FOUR: Convert datasets to json format and export to MongoDB using terminal

```python
# Convert dataframe to json format
# Write data to MongoDB 
organized_df1 = organized_df.to_json(orient='records')
with open('organized_df1.json', 'w') as f:
    f.write(organized_df1)
```

```python
# Convert dataframe to json format
# Write data to MongoDB 
recipes_df1 = recipes_df.to_json(orient='records')
with open('recipes_df1.json', 'w') as f:
    f.write(recipes_df1)
```
